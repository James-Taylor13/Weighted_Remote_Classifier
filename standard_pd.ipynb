```python
import pandas as pd

from jobs_classifier.text_preprocessing.preprocessing import Preprocess
from jobs_classifier import model_setup
from jobs_classifier import enums


def inference_weighted(text1, weight1=1, text2=None, weight2=None, text3=None, weight3=None, print_out=True):
    if type(text1) == str:
        text1 = [text1]

    clean_text1 = [Preprocess(t).text.lower() for t in text1]

    tfidf_module = model_setup.tfidf_trainer(pretrained_model=True)
    tfidf_out = tfidf_module.transform(clean_text1)

    model = model_setup.JobSummaryClassifier(load_pretrained=True)
    predictions = model.predict(tfidf_out)[:, 1]

    output_summary = pd.DataFrame({"Text": text1, "Confidence": list(predictions)})
    output_summary["Class"] = output_summary["Confidence"] >= enums.THRESHOLD

    output_summary.to_csv("model_predictions/inference.csv", index=False)

    if print_out:
        for ix, (t, pred) in enumerate(zip(text1, predictions)):
    
            if pred >= enums.THRESHOLD:
                if pred >= enums.HIGH_CONFIDENCE:
                    print(f"\n{t} -------------> \n [GREEN]: Location Negotiable EGLIBLE with high confidence")
                    print('PRED:' + str(pred))

                else: 
                    print(f"\n{t} -------------> \n [YELLOW]: Location Negotiable EGLIBLE medium confidence")
                    print('PRED:' + str(pred))
            else:
                if 1-pred >= enums.HIGH_CONFIDENCE:
                    print(f"\n{t} -------------> \n [RED]: Location Negotiable INEGLIBLE with high confidence")
                    print('PRED:' + str(pred))
                else: 
                    print(f"\n{t} -------------> \n [YELLOW]: Location Negotiable INEGLIBLE with medium confidence")
                    print('PRED:' + str(pred))
        weighted_confidence1 = pred * weight1

    if text2 == None:
        final_confidence = weighted_confidence1
        return print("FINAL CONFIDENCE:" + str(final_confidence))
    else:
        if type(text2) == str:
            text2 = [text2]

        clean_text2 = [Preprocess(t).text.lower() for t in text2]

        tfidf_module = model_setup.tfidf_trainer(pretrained_model=True)
        tfidf_out = tfidf_module.transform(clean_text2)

        model = model_setup.JobSummaryClassifier(load_pretrained=True)
        predictions = model.predict(tfidf_out)[:, 1]

        output_summary = pd.DataFrame({"Text": text2, "Confidence": list(predictions)})
        output_summary["Class"] = output_summary["Confidence"] >= enums.THRESHOLD

        output_summary.to_csv("model_predictions/inference.csv", index=False)

    if print_out:
        for ix, (t, pred) in enumerate(zip(text2, predictions)):
    
            if pred >= enums.THRESHOLD:
                if pred >= enums.HIGH_CONFIDENCE:
                    print(f"\n{t} -------------> \n [GREEN]: Location Negotiable EGLIBLE with high confidence")
                    print('PRED:' + str(pred))
                else: 
                    print(f"\n{t} -------------> \n [YELLOW]: Location Negotiable EGLIBLE medium confidence")
                    print('PRED:' + str(pred))
            
            else:
                if 1-pred >= enums.HIGH_CONFIDENCE:
                    print(f"\n{t} -------------> \n [RED]: Location Negotiable INEGLIBLE with high confidence")
                    print('PRED:' + str(pred))

                else: 
                    print(f"\n{t} -------------> \n [YELLOW]: Location Negotiable INEGLIBLE with medium confidence")
                    print('PRED:' + str(pred))
        weighted_confidence2 = pred * weight2
        
    if text3 == None:
        final_confidence = weighted_confidence1 + weighted_confidence2
        return print("FINAL CONFIDENCE:" + str(final_confidence))
    else:
        if type(text3) == str:
            text3 = [text3]

        clean_text3 = [Preprocess(t).text.lower() for t in text3]

        tfidf_module = model_setup.tfidf_trainer(pretrained_model=True)
        tfidf_out = tfidf_module.transform(clean_text3)

        model = model_setup.JobSummaryClassifier(load_pretrained=True)
        predictions = model.predict(tfidf_out)[:, 1]

        output_summary = pd.DataFrame({"Text": text3, "Confidence": list(predictions)})
        output_summary["Class"] = output_summary["Confidence"] >= enums.THRESHOLD

        output_summary.to_csv("model_predictions/inference.csv", index=False)

        if print_out:
            for ix, (t, pred) in enumerate(zip(text3, predictions)):

                if pred >= enums.THRESHOLD:
                    if pred >= enums.HIGH_CONFIDENCE:
                        print(f"\n{t} -------------> \n [GREEN]: Location Negotiable EGLIBLE with high confidence")
                        print('PRED:' + str(pred))
                    else: 
                        print(f"\n{t} -------------> \n [YELLOW]: Location Negotiable EGLIBLE medium confidence")
                        print('PRED:' + str(pred))
                
                else:
                    if 1-pred >= enums.HIGH_CONFIDENCE:
                        print(f"\n{t} -------------> \n [RED]: Location Negotiable INEGLIBLE with high confidence")
                        print('PRED:' + str(pred))
                    else: 
                        print(f"\n{t} -------------> \n [YELLOW]: Location Negotiable INEGLIBLE with medium confidence")
                        print('PRED:' + str(pred))
            weighted_confidence3 = pred * weight3

    final_confidence = weighted_confidence1 + weighted_confidence2 + weighted_confidence3
    print("FINAL CONFIDENCE:" + str(final_confidence))
```
